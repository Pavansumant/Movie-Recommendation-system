```python
import numpy as np
import pandas as pd
```

```python
movies = pd.read_csv('/kaggle/input/movies/movies.dat' , sep='::' , engine='python' , header=None)
```

```python
movies.head()
```

```python
movies[2][0].split('|')
```

```python
genres  = []
```

```python
for i in movies[2]:
  temp= i.split('|')
  temp = ' '.join(temp)
  genres.append(temp)
```

```python
genres
```

```python
movies[1][0].split(' (')
```

```python
movi = []

for i in movies[1]:
  temp = i.split(' (')
  temp=''.join(temp[0])
  movi.append(temp)
```

```python
movi
```

```python
movies['Movie' ] = pd.DataFrame(movi)
```

```python
movies['Genres'] = pd.DataFrame(genres)
```

```python
movies.head()
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vectorizer = TfidfVectorizer()
X = vectorizer.fit_transform(movies['Genres'])
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(X , X)
```

```python
X.shape
```

```python
similarity.shape
```

```python
movie_title = movies['Movie']
```

```python
indices = pd.Series(movies.index , movie_title)
```

```python
indices
```

```python
def movie_recommender(title):
  index = indices[title]
  similarity_score = list(enumerate(similarity[index]))
  similarity_score = sorted(similarity_score , key=lambda x:x[1] , reverse=True)
  similarity_score = similarity_score[1:31]
  movie_indices = [i[0] for i in similarity_score]
  return movie_title.iloc[movie_indices]
```

```python
movie_recommender('Toy Story')
```

```python
movie_recommender('Atlantis: The Lost Empire')
```
